```python
df=pd.read_csv(r'C:\Users\Khuzaima Aziz\Downloads\orders.csv.csv', encoding='latin-1')
```

```python
data = df.drop_duplicates()
data = df.dropna()
```

```python
data=data.loc[:,'air fresheners candles':]
```

```python
data
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(data)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2)
kmeans.fit(data_scaled)
```

```python
clusters = kmeans.labels_

```

```python
df['gender']=clusters
```

```python
df['gender'].value_counts()
```

```python
cluster_totals = df.groupby('gender').sum().iloc[:, :-1]
```

```python
cluster_totals['feminine care']
```

```python
df['gender'].replace(1,'Female',inplace=True)
```

```python
df['gender'].replace(0,'Male',inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
features=df.drop('gender',axis=1)
labels=df['gender']
```

```python
X_train,X_test,y_train,y_test=train_test_split(features, labels, test_size=0.05)
```
