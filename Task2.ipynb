```python
import pandas as pd
pd.set_option('display.max_columns', None)
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r'C:\Users\Khuzaima Aziz\Downloads\orders.csv.csv', encoding='latin-1')
```

```python
data = df.drop_duplicates()
data = df.dropna()
```

```python
data=data.loc[:,'air fresheners candles':]
```

```python
data
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(data)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2)
kmeans.fit(data_scaled)
```

```python
clusters = kmeans.labels_

```

```python
df['gender']=clusters
```

```python
df['gender'].value_counts()
```

```python
cluster_totals = df.groupby('gender').sum().iloc[:, :-1]
```

```python
cluster_totals['feminine care']
```
